```python
# Join the last two words back into a regular text string
df['Clean_Location'] = df['Location'].str.split().str[-2:].str.join(' ')

# Look at the first 10 rows to make sure the brackets are gone
df['Clean_Location'].head(10)
```

```python
states = { 'andhra pradesh', 'arunachal pradesh', 'assam', 'bihar', 'chhattisgarh', 'goa', 'gujarat', 'haryana', 'himachal pradesh', 'jharkhand', 'karnataka', 'kerala', 'madhya pradesh', 'maharashtra', 'manipur', 'meghalaya', 'mizoram', 'nagaland', 'odisha', 'punjab', 'rajasthan', 'sikkim', 'tamil nadu', 'telangana', 'tripura', 'uttar pradesh', 'uttarakhand', 'west bengal', 'delhi'}
```

```python
import numpy as np

def extract_state(location):
    if pd.isna(location):
        return np.nan

    location = str(location).lower()

    for state in states:
        if state in location:
            return state.title()

    return np.nan
```

```python
df['State'] = df['Clean_Location'].apply(extract_state)
```

```python
print("Missing States:", df['State'].isnull().sum())
```

```python
# 1. Look at the first word (which is usually the City) in our clean location
df['City_Guess'] = df['Clean_Location'].str.split().str[0]

# 2. Build a map of City -> State from the rows that actually have a State
state_map = df.dropna(subset=['State']).drop_duplicates('City_Guess').set_index('City_Guess')['State'].to_dict()

# 3. If any famous cities are missing from your map, we manually add them here to be safe
state_map['Agartala'] = 'Tripura'
state_map['Gurugram'] = 'Haryana'
state_map['Ahmadnagar'] = 'Maharashtra'
state_map['Noida'] = 'Uttar Pradesh'

# 4. Fill in the missing States using our map!
df['State'] = df['State'].fillna(df['City_Guess'].map(state_map))

# 5. Fill any absolute leftovers with 'Other' so the model doesn't crash
df['State'] = df['State'].fillna('Other')
```

```python
# Show 30 random raw locations where the State ended up as 'Other'
df[df['State'] == 'Other']['Location'].sample(30, random_state=42)
```

```python
# Expanded list covering NCR regions, Haryana, UP, Gujarat, Goa, and more
rescue_map = {
    # Delhi & NCR / Haryana
    'delhi': 'Delhi', 'new delhi': 'Delhi', 'dwarka': 'Delhi', 'rohini': 'Delhi',
    'gurugram': 'Haryana', 'gurgaon': 'Haryana', 'faridabad': 'Haryana', 
    'sonipat': 'Haryana', 'panchkula': 'Haryana', 'rohtak': 'Haryana', 'yamuna nagar': 'Haryana',
    
    # Uttar Pradesh / Noida NCR
    'noida': 'Uttar Pradesh', 'greater noida': 'Uttar Pradesh', 'ghaziabad': 'Uttar Pradesh', 
    'vaishali': 'Uttar Pradesh', 'kanpur': 'Uttar Pradesh', 'gorakhpur': 'Uttar Pradesh', 
    'meerut': 'Uttar Pradesh', 'barra': 'Uttar Pradesh', 'mirzapur': 'Uttar Pradesh',
    
    # Gujarat
    'ahmedabad': 'Gujarat', 'surat': 'Gujarat', 'vadodara': 'Gujarat', 
    'gandhinagar': 'Gujarat', 'anand': 'Gujarat', 'nadiad': 'Gujarat', 'ode': 'Gujarat',
    
    # Goa
    'mapusa': 'Goa', 'chimbel': 'Goa', 'miramar': 'Goa',
    
    # Other Major States (Carried Over)
    'bengaluru': 'Karnataka', 'bangalore': 'Karnataka', 'mysuru': 'Karnataka',
    'hyderabad': 'Telangana', 'secunderabad': 'Telangana',
    'chennai': 'Tamil Nadu', 'kolkata': 'West Bengal',
    'jaipur': 'Rajasthan', 'jodhpur': 'Rajasthan', 'kota': 'Rajasthan',
    'kochi': 'Kerala', 'cochin': 'Kerala', 'mumbai': 'Maharashtra', 
    'pune': 'Maharashtra', 'thane': 'Maharashtra', 'kalyan': 'Maharashtra',
    'chandigarh': 'Chandigarh', 'bhubaneswar': 'Odisha', 'rudrapur': 'Uttarakhand',
    'ludhiana': 'Punjab', 'bathinda': 'Punjab', 'dehradun': 'Uttarakhand'
}

# 2. Define the rescue function (Same as before)
def rescue_other_states(row):
    if row['State'] != 'Other' and pd.notna(row['State']):
        return row['State']
    
    location_text = str(row['Location']).lower()
    for keyword, correct_state in rescue_map.items():
        if keyword in location_text:
            return correct_state
            
    return 'Other'

# 3. Apply the updated map
df['State'] = df.apply(rescue_other_states, axis=1)

# 4. Check the results!
print(df['State'].value_counts().head(10))
```
